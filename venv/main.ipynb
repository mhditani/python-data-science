```python
X = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]
```

```python
plt.plot(X, y)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
# subplot has fig & axis
fig, ax = plt.subplots()
ax.plot(X, y)
```

```python
X = [22, 88, 98, 103, 45]
y = [7, 8, 9, 10, 11]

fig, ax = plt.subplots()   # <-- lowercase s

ax.plot(X, y)
ax.set(title="Sample Plot", xlabel="X label", ylabel="Y label")

fig.savefig("myfigure.png")
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = x**2
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set(title="Sample Plot 2", xlabel="X label", ylabel="Y label")
fig.savefig("SamplePlot2.png")
plt.show()
```

```python
favorite_foods = {
    "Almond Butter": 10,
    "Blueberries" : 5,
    "Eggs": 6
}

fig, ax = plt.subplots()
ax.bar(favorite_foods.keys(), favorite_foods.values())
ax.set(title="Favorite Foods", xlabel="Food Name", ylabel="Food Price")
fig.savefig("favfoods.png")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.barh(list(favorite_foods.keys()), list(favorite_foods.values()))
ax.set(title="Favorite Foods", xlabel="Food Name", ylabel="Food Price")
plt.show()
```
